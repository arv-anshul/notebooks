### Making pipeline

## Random Forest Classifier

```python
# define columns for ordinal encoding
ordinal_col = ['EDUCATION'] 

# define columns for one hot encoding
cat_col = ['MARITALSTATUS', 'GENDER', 'last_prod_enq2','first_prod_enq2']

#define columns for standardization
num_col=feature.select_dtypes('number').columns

# Define the pipeline for numerical features
num_pipe = Pipeline(steps=[
    ('scale', StandardScaler())
])

# Define the pipeline for categorical features
cat_pipe = Pipeline(steps=[ # (name, transform)
    ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
])

# Define the pipeline for ordinal features
ordi_pipe = Pipeline(steps=[
    ('ordinal', OrdinalEncoder(categories=[['SSC', '12TH', 'OTHERS', 'GRADUATE', 'UNDER GRADUATE', 'PROFESSIONAL', 'POST-GRADUATE']]))
])

# Combine the transformers in a ColumnTransformer
process = ColumnTransformer(transformers=[ #(name, transformer, columns)
    ('num_tran', num_pipe, num_col),
    ('cat_tran', cat_pipe, cat_col),
    ('ori_tran', ordi_pipe, ordinal_col)
], remainder='passthrough')

# Define the final pipeline with the preprocessing and model
final_rf = Pipeline([
    ('process', process),
    ('model', RandomForestClassifier())
])

# Fit the final pipeline
final_rf.fit(x_train, y_train)
```

```python
y_pred_rf=final_rf.predict(x_test)
y_pred_rf
```

```python
def evaluation(actual,pre):
    acc=accuracy_score(actual,pre)
    pr=precision_score(actual,pre,average='macro')
    re=recall_score(actual,pre,average='macro')
    f_score=f1_score(actual,pre,average='macro')
    c_f=confusion_matrix(actual,pre)
    
    return {
        "accuracy":acc,
        "pre":pr,
        "recal":re,
        "f_score":f_score,
        "matrix":c_f
    }
```

```python
evaluation(actual = y_test,pre= y_pred_rf)
```

```python
evaluation(y_test,y_pred_rf)['matrix']
```

```python
from sklearn.metrics import classification_report
# Generate classification report
report = classification_report(y_test, y_pred_rf)

print(report)
```

```python
precision, recall, f1_score, _ = precision_recall_fscore_support(y_test, y_pred_rf)

for i, v in enumerate(['p1', 'p2', 'p3', 'p4']):
    print(f"Class {v}:")
    print(f"Precision: {precision[i]}")
    print(f"Recall: {recall[i]}")
    print(f"F1 Score: {f1_score[i]}")
    print()
    
```

```python
x_train,x_test,y_train,y_test=train_test_split(feature,label,test_size=0.2,random_state=42)
```

```python

categorical_cols = ['MARITALSTATUS', 'GENDER', 'last_prod_enq2', 'first_prod_enq2', 'EDUCATION']

oe = OneHotEncoder(handle_unknown='ignore', drop='first')
X_train_encoded = oe.fit_transform(x_train[categorical_cols])
X_test_encoded = oe.transform(x_test[categorical_cols])

# Convert the sparse matrix to a dense array and assign it to the DataFrame column
X_train_encoded_df = pd.DataFrame(X_train_encoded.toarray(), columns=oe.get_feature_names_out(categorical_cols), index=x_train.index)
X_test_encoded_df = pd.DataFrame(X_test_encoded.toarray(), columns=oe.get_feature_names_out(categorical_cols), index=x_test.index)

# Drop the original 'GENDER' column from the DataFrames
x_train.drop(columns=categorical_cols, inplace=True)
x_test.drop(columns=categorical_cols, inplace=True)

# Concatenate the encoded columns to the original DataFrames
X_train = pd.concat([x_train, X_train_encoded_df], axis=1)
X_test = pd.concat([x_test, X_test_encoded_df], axis=1)
```

```python
from skopt import BayesSearchCV

# Define the hyperparameter search space
param_space = {
    'colsample_bytree': (0.1, 0.9),
    'learning_rate': (0.001, 1.0, 'log-uniform'),
    'max_depth': (3, 10),
    'alpha': (1, 100),
    'n_estimators': (10, 100)
}

# Initialize the XGBoost classifier
model = XGBClassifier(objective='multi:softmax', num_class=4)

# Create BayesSearchCV object
bayes_search = BayesSearchCV(model, param_space, n_iter=50, cv=5, scoring='accuracy', random_state=42)

# Encode the target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y_train)

# Perform Bayesian optimization on the training set
bayes_search.fit(X_train, y_train)

# Get the best model
best_model = bayes_search.best_estimator_

# Predict on test set
y_pred = best_model.predict(X_test)

# Calculate accuracy
test_accuracy = accuracy_score(y_test, y_pred)

print(f"Best hyperparameters: {bayes_search.best_params_}")
print(f"Test accuracy: {test_accuracy:.2f}")
```
